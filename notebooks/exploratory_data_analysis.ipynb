```python
# imports
%matplotlib inline

import matplotlib.pyplot as plt
import sys
import os
import pandas as pd
import numpy as np
import warnings
import time
import threading
import random
import sklearn.linear_model as sk


sys.path.append(os.path.abspath('../'))
from src.utils.address_utils import *

address = Address(prefix='../')
```

```python
# Reading data
df_user = pd.read_csv(address.data.handles)
df_problem = pd.read_csv(address.data.problems).fillna(-1)
```

```python
# Analyzing handles.csv
ax_rating_count = plt.axes()
ax_rating_count.hist(df_user['maxRating'], bins=15)
ax_rating_count.set_xlabel('Rating')
ax_rating_count.set_ylabel('Number of users')

print(f'Average rating {np.mean(df_user["maxRating"])}')
print(f'Median rating {np.median(df_user["maxRating"])}')
print(f'Std Dev rating {np.std(df_user["maxRating"])}')
```

```python
# Analyzing problems.csv

ax_problem_rating_num_solved = plt.axes()
ax_problem_rating_num_solved.scatter(df_problem['rating'],df_problem['solvedCount']);
```

```python
# Analyzing size of submission file, size no of files read in 30 seconds interval graph

file = open(address.log.problem_diff, mode = 'r')
lines = file.readlines()
file.close()
user_val = []
time_val = []
try:
    for line in lines:
        user_val.append(int(line.split('/')[0]))
        time_val.append(float(line.split(':')[-1][:-1]))
except:
    pass
ax_submission_size = plt.axes()
ax_submission_size.plot(user_val, time_val)
ax_submission_size.set_xlabel('num_users')
ax_submission_size.set_ylabel('time taken');
```

```python
# Analyzing problem difficulty

df_difficulty = pd.read_csv(address.data.problem_diff)

ax_difficulty_rating = plt.axes()
ax_difficulty_rating.scatter(df_difficulty['difficulty'], df_problem['rating'])
ax_difficulty_rating.set_xlabel('difficulty')
ax_difficulty_rating.set_ylabel('rating');
```

```python
ax_difficulty_solvedct = plt.axes()
ax_difficulty_solvedct.scatter(df_difficulty['difficulty'], df_problem['solvedCount'])
ax_difficulty_solvedct.set_xlabel('difficulty')
ax_difficulty_rating.set_ylabel('solved count');
```
