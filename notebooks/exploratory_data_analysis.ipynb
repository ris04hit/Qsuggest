```python
# imports
%matplotlib inline

import matplotlib.pyplot as plt
import sys
import os
import pandas as pd
import numpy as np
```

```python
# Reading data
df_user = pd.read_csv('../data/scraped/handles.csv')
```

```python
# Analyzing handles.csv
ax_rating_count = plt.axes()
ax_rating_count.hist(df_user['maxRating'], bins=15)
ax_rating_count.set_xlabel('Rating')
ax_rating_count.set_ylabel('Number of users')

print(f'Average rating {np.mean(df_user["maxRating"])}')
print(f'Median rating {np.median(df_user["maxRating"])}')
print(f'Std Dev rating {np.std(df_user["maxRating"])}')
```
